```python
#Read the file from the link givcen below:
#https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-
#Learning-and-Natural-Language-Processing.docx (it has been uplaoded to google colab contents folder)
#Read and display first 5 paragraphs from it.
#Count total ‘The’ present in the text.
```

```python
# import the class
from spire.doc import Document
```

```python
# create the object
document = Document()
```

```python
# load the file
document.LoadFromFile('Machine-Learning-and-Natural-Language-Processing.docx')
```

```python
document_text = document.GetText()
```

```python
print(document_text)
```

```python
# Get the first section in the document
section = document.Sections[0]
```

```python
# Get the paragraph from section
paragraph = section.Paragraphs[1]
paragraph.Text
```

```python
from spire.doc import FixedLayoutDocument
```

```python
layoutDoc = FixedLayoutDocument(document)
```

```python
# retrieve the pages
layoutPage = layoutDoc.Pages[0]
```

```python
print(layoutPage.Text)
```

```python
len(layoutDoc.Pages)
```

```python
count = 1
for page in layoutDoc.Pages:
    try:
        print('\t\t\t\t\tPage:', count)
        print(page.Text)
        count += 1
    except Exception:
        break
```

```python
if len(layoutDoc.Pages) < 2:
    print("Error: There are not enough pages in the document.")
else:
    page = layoutDoc.Pages[1]
```
