# Abstractive Summarization

Download the data from the link below using web scrapping:
https://en.wikipedia.org/wiki/Sachin_Tendulkar
Apply the abstractive summarization to summarize this text.

```python
#from transformers import pipeline
```

```python
!pip install torch
!pip install transformers
from transformers import pipeline
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Sachin_Tendulkar'
page =urlopen(url)
html_text = page.read().decode('utf-8')

# Use Beautiful Soup to parse the HTML content
soup = BeautifulSoup(html_text, 'html.parser')
#txt = soup.get_text()

# Extract all the text from the <p> (paragraph) tags
paragraphs = [p.get_text() for p in soup.find_all('p')]

# Join the paragraphs into a single string
text_content = '\n'.join(paragraphs)

# Initialize the summarization pipeline with a more robust model
#text_summarizer = pipeline('summarization', model="facebook/bart-large-cnn") # Use a more robust model
text_summarizer = pipeline('summarization')

# Truncate the input text to a reasonable length
max_input_length = 1024  # Adjust this based on the model's limitations
truncated_text = text_content[:max_input_length]

# Now pass the cleaned and truncated text content to the summarizer
#output = text_summarizer(truncated_text, max_length=200, min_length=10)
output = text_summarizer(truncated_text, max_length=200, min_length=10)

# Print the summarized output
print(output[0]['summary_text'])
```

# bag of words using count vectorizer.

4. Use following sentences to find the bag of words using count vectorizer.
1. The postman delivered the package to the wrong address.
2. I wrapped a beautiful present for my friend's birthday.
3. The delivery truck arrived late due to heavy traffic.
4. We need to check the shipping address before sending the order.
5. Online shopping offers a wide variety of products with fast delivery.

```python
sent1 = 'The postman delivered the package to the wrong address.'
sent2 = 'I wrapped a beautiful present for my friend\'s birthday.'
sent3='The delivery truck arrived late due to heavy traffic.'
sent4='We need to check the shipping address before sending the order.'
sent5='Online shopping offers a wide variety of products with fast delivery.'
```

```python
import pandas as pd
```

```python
# import the count vectorizer class
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# create the object
cv = CountVectorizer()
```

```python
# apply the BOW method
mydata = cv.fit_transform([sent1, sent2, sent3, sent4, sent5])
```

```python
mydata.toarray()
```

```python
cv.get_feature_names_out()
```

```python
df = pd.DataFrame(data=mydata.toarray(),
                  columns=cv.get_feature_names_out())
```
